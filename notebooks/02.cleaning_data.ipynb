```python
# Data manipulation
import pandas as pd
pd.set_option('display.max_columns', None)

# System paths for module importing
import sys 
sys.path.append('../')

# Custom utility functions
from src import data_cleaning_utils as dcu
```

```python
df_raw=pd.read_csv("../data/data_raw.csv")
df_raw.sample(5)
```

```python
df=df_raw.copy()
```

## Standardize column names

```python
df.columns=df.columns.str.replace(" ","_").str.lower()
```

## Convert visit_date column to datetime format

```python
df['visit_date']=pd.to_datetime(df['visit_date'],format='%Y-%m-%d')
```

## Create human-readable Conversion string column

```python
df['conversion_str']=df['conversion'].apply(lambda x: 'yes' if x==1 else 'no')
```

```python
df.loc[:,['conversion','conversion_str']].sample(5)
```

## Replace spaces with underscores

```python
dcu.replace_spaces_with_underscores(df)
```

## Standarize decimal separators and convert to float

```python
dcu.convert_commas_to_dots(df)
```

## Convert all string columns to lowercase

```python
dcu.lowercase_strings(df)
```

```python
df.sample(5)
```

## Clean Product Category: Fix 'FASHHION' typo to 'Fashion'

```python
df['product_category']=df['product_category'].str.replace('fashhion','fashion')
```

```python
df.sample(5)
```

## Clean Product id: Strip 'P' prefix and convert to numeric

```python
df['product_id']=df['product_id'].str.replace('p','')
df['product_id']=df['product_id'].astype('Int64')
```

## Clean Browser Column using a descriptive map

```python
df['browser'].unique()
```

```python
browser_clean_map={'safaaaaari': 'safari',
         'ottttther':'other',
         'firefoxxxxxx':'firefox',
        'eddddge': 'edge',
         'chrooome': 'chrome' }
```

```python
df['browser']=df['browser'].map(browser_clean_map)
```

```python
df.to_csv('../data/data_cleaned.csv',index=False)
```
